SOAL 3: MODELING & EVALUATION - Framework SHAP

Muat Model & Aset

```python
import joblib
import pandas as pd
import matplotlib.pyplot as plt
import shap

# Muat data uji dan model terbaik dari folder lokal
_, X_test, _, _, _ = joblib.load('../data/processed/clean_data.pkl')
best_model = joblib.load('../models/best_model.pkl')
print("✓ Model dan data uji siap dibedah dengan SHAP.")
```

Kalkulasi Nilai SHAP

```python
# Inisialisasi TreeExplainer
explainer = shap.TreeExplainer(best_model)

# Mengambil sampel acak 300 data uji agar kalkulasi di komputer lokal berjalan ringan dan cepat
X_sample = X_test.sample(n=300, random_state=42)
shap_values = explainer(X_sample)

print("✓ Kalkulasi nilai SHAP untuk 300 data sampel selesai.")
```

Visualisasi SHAP Summary Plot & Ekspor Gambar

```python
plt.figure(figsize=(10, 6))

# Membuat Summary Plot (Beeswarm)
shap.summary_plot(shap_values, X_sample, show=False)
plt.title('SHAP Summary Plot: Faktor Penentu Utama Emisi CO2', fontsize=14, pad=20)
plt.tight_layout()

# Simpan otomatis gambar ke folder reports
plt.savefig('../reports/shap_summary.png')
plt.show()
```
